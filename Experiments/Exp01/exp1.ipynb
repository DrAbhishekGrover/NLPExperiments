```python
#Import Libraries
import nltk
from nltk.corpus import movie_reviews, stopwords
nltk.download('movie_reviews')
nltk.download('stopwords')

stop_words=set(stopwords.words('english'))
#print(stop_words)
doc=[" ".join(movie_reviews.words(fid)) for fid in movie_reviews.fileids()]
label=[movie_reviews.categories(id)[0] for id in movie_reviews.fileids()]
print(len(doc))
print(len(label))
print(doc[1])
print(label[1])
```

```python
import re
from nltk.stem import PorterStemmer,WordNetLemmatizer
nltk.download('movie_reviews')
nltk.download('stopwords')

#wordnet lemmatizer
#punkt for tokenization
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')


stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()

def clean_texta(text):
    text=text.lower()#Normalization
    text=re.sub(r'[^a-z\s]','',text)
    tokens=nltk.word_tokenize(text)
    tokens=[lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)

def clean_textb(text):
    text=text.lower()#Normalization
    text=re.sub(r'[^a-z\s]','',text)
    tokens=nltk.word_tokenize(text)
    tokens=[stemmer.stem(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)

print(f'Raw Text:{doc[1]}')
print(f'After Lemmatization:{clean_texta(doc[1])}')
print(f'After Stemming:{clean_textb(doc[1])}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

doc=[" ".join(movie_reviews.words(fid)) for fid in movie_reviews.fileids()]
#doc=[clean_texta(" ".join(movie_reviews.words(fid))) for fid in movie_reviews.fileids()]
#doc=[clean_textb(" ".join(movie_reviews.words(fid))) for fid in movie_reviews.fileids()]
label=[movie_reviews.categories(id)[0] for id in movie_reviews.fileids()]
X_train,X_test,Y_train,Y_test=train_test_split(doc,label,test_size=0.2,random_state=5)


#To generate Bag of Words Vectors
v=CountVectorizer(max_features=5000)
X_trainv=v.fit_transform(X_train)
X_testv=v.transform(X_test)
#print(X_train[1])
#print(X_test[1])

mymod=MultinomialNB()
mymod.fit(X_trainv,Y_train)
#print(f'Review:{X_train[8]}')
#print(f'Original Label:{Y_train[8]}')
#print(f'Predicted label:{mymod.predict(X_trainv[8].reshape(1,-1))}')

Ytestpred=mymod.predict(X_testv)
Ytrainpred=mymod.predict(X_trainv)
print(f'Train Accuracy:{accuracy_score(Y_train,Ytrainpred)}')
print(f'Test Accuracy:{accuracy_score(Y_test,Ytestpred)}')
```
